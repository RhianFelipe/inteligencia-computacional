```python
import math
import matplotlib.pyplot as plt

# Definindo os dados de treinamento: Temperatura, Umidade e a Classe (Flor)
dados_treinamento = [
    (10, 20, 'Classe 1'), (8, 13, 'Classe 0'), (6, 15, 'Classe 1'),
    (5, 11, 'Classe 1'), (20, 26, 'Classe 0'), (22, 26, 'Classe 0'),
    (23, 27, 'Classe 1'), (18, 23, 'Classe 0')
]

# Definindo os pontos a serem classificados: c1 e c2
c1 = (5, 10)
c2 = (25, 20)

k = 3  # Valor de k (número de vizinhos mais próximos)
```

```python
def distancia_euclidiana(ponto1, ponto2):
    return math.sqrt((ponto1[0] - ponto2[0])**2 + (ponto1[1] - ponto2[1])**2)
```

```python
def encontrar_vizinhos(ponto, dados, k):
    distancias = []
    for dado in dados:
        dist = distancia_euclidiana(ponto, (dado[0], dado[1]))
        distancias.append((dist, dado[2]))  # Armazena a distância e a classe

    distancias.sort()  # Ordena as distâncias
    vizinhos = distancias[:k]  # Seleciona os k vizinhos mais próximos
    return vizinhos
```

```python
vizinhos_c1 = encontrar_vizinhos(c1, dados_treinamento, k)
vizinhos_c2 = encontrar_vizinhos(c2, dados_treinamento, k)

print("Vizinhos mais próximos de c1:", vizinhos_c1)
print("Vizinhos mais próximos de c2:", vizinhos_c2)
```

```python

# Separando os dados por classes para plotagem
classe0 = [(d[0], d[1]) for d in dados_treinamento if d[2] == 'Classe 0']
classe1 = [(d[0], d[1]) for d in dados_treinamento if d[2] == 'Classe 1']

# Plotando os pontos de treinamento e os pontos c1 e c2
plt.scatter(*zip(*classe0), color='red', label='Classe 0')
plt.scatter(*zip(*classe1), color='blue', label='Classe 1')
plt.scatter(c1[0], c1[1], color='black', marker='x', label='c1')
plt.scatter(c2[0], c2[1], color='orange', marker='x', label='c2')

plt.xlabel('Temperatura')
plt.ylabel('Umidade')
plt.legend()
plt.title('Pontos de Treinamento e Pontos c1 e c2')

plt.show()
```
